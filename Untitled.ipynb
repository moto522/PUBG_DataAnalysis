```python
import numpy as np
import pandas as pd
pubg_death = pd.read_csv('kill_match_stats_final_3.csv')
pubg_death.head()
```

```python
# 必要ない列を削除
pubg_death.drop(columns=['killer_name'],inplace=True)
pubg_death.drop(columns=['match_id'],inplace=True)
pubg_death.drop(columns=['victim_name'],inplace=True)
pubg_death
```

```python
# NaNを削除
pubg_death.dropna(inplace=True)
```

```python
pubg_death2=pubg_death[pubg_death['killer_position_x']!=0.0]
pubg_death2=pubg_death[pubg_death['victim_position_x']!=0.0]
pubg_death2
```

```python
pubg_death['distance'] = (np.sqrt((pubg_death['victim_position_x']-pubg_death['killer_position_x'])**2+(pubg_death['victim_position_y']-pubg_death['killer_position_y'])**2))*0.02
pubg_death
```
